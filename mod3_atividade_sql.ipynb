Firstly, i created a bucket and make a table.
```sql
CREATE TABLE EXTERNAL transacoes(
  id_cliente BIGINT,
  id_transacao BIGINT,
  valor FLOAT,
  id_loja STRING
)
ROW FORMAT SERDE 'org.apache.hadoop.hive.serde2.OpenCSVSerde'
WITH SERDEPROPERTIES (
  'separatorChar' = ',',
  'quoteChar' = '"',
  'escapeChar' = '\\'
)
STORED AS TEXTFILE
LOCATION 's3://mod3-jeduardo/';
```

### **Query 1**


### **Query 2**

```sql
SELECT id_cliente, valor, id_loja AS nome_loja
```

### **Query 3**

```sql

FROM transacoes;
```

````python
query_3
````

### **Query 4**

```sql
SELECT id_cliente, valor
FROM transacoes
ORDER BY valor DESC
LIMITE 2;
```
